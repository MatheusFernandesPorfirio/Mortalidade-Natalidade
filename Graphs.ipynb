# Import das bibliotecas necessárias e das bases de dados

```python
!pip install matplotlib
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import requests
import io

#Importação dos dados de Mortalidade
url = "https://raw.githubusercontent.com/MatheusFernandesPorfirio/Mortalidade-Natalidade/main/MORTALIDADE.csv"
download = requests.get(url).content
df1 = pd.read_csv(io.StringIO(download.decode('utf-8')))

#Importação dos dados de Natalidade
url = "https://raw.githubusercontent.com/MatheusFernandesPorfirio/Mortalidade-Natalidade/main/NATALIDADE.csv"
download = requests.get(url).content
df2 = pd.read_csv(io.StringIO(download.decode('utf-8')))

#Fonte do trabalho:
font = {'family':'serif','color':'black','size':15}
```

```python
#Divisão dos anos da base de Mortalidade
ano1m = df1[df1.ano==2016]
ano2m = df1[df1.ano==2017]
ano3m = df1[df1.ano==2018]
ano4m = df1[df1.ano==2019]
ano5m = df1[df1.ano==2020]

#Divisão dos anos da base de Natalidade
ano1n = df2[df2.ano==2016]
ano2n = df2[df2.ano==2017]
ano3n = df2[df2.ano==2018]
ano4n = df2[df2.ano==2019]
ano5n = df2[df2.ano==2020]
```

#Gráfico de linha sobre a mortalidade e natalidade ao longo dos anos

```python
#Busca de todos os indivíduos mortos e nascidos de 2016
masculino_m1 = ano1m['sexo_masculino'].sum()
feminino_m1 = ano1m['sexo_feminino'].sum()
sexo_ignorado_m1 = ano1m['sexo_ignorado'].sum()
masculino_n1 = ano1n['sexo_masculino'].sum()
feminino_n1 = ano1n['sexo_feminino'].sum()
sexo_ignorado_n1 = ano1n['sexo_ignorado'].sum()

#Busca de todos os indivíduos mortos e nascidos de 2017
masculino_m2 = ano2m['sexo_masculino'].sum()
feminino_m2 = ano2m['sexo_feminino'].sum()
sexo_ignorado_m2 = ano2m['sexo_ignorado'].sum()
masculino_n2 = ano2n['sexo_masculino'].sum()
feminino_n2 = ano2n['sexo_feminino'].sum()
sexo_ignorado_n2 = ano2n['sexo_ignorado'].sum()

#Busca de todos os indivíduos mortos e nascidos de 2018
masculino_m3 = ano3m['sexo_masculino'].sum()
feminino_m3 = ano3m['sexo_feminino'].sum()
sexo_ignorado_m3 = ano3m['sexo_ignorado'].sum()
masculino_n3 = ano3n['sexo_masculino'].sum()
feminino_n3 = ano3n['sexo_feminino'].sum()
sexo_ignorado_n3 = ano3n['sexo_ignorado'].sum()

#Busca de todos os indivíduos mortos e nascidos de 2019
masculino_m4 = ano4m['sexo_masculino'].sum()
feminino_m4 = ano4m['sexo_feminino'].sum()
sexo_ignorado_m4 = ano4m['sexo_ignorado'].sum()
masculino_n4 = ano4n['sexo_masculino'].sum()
feminino_n4 = ano4n['sexo_feminino'].sum()
sexo_ignorado_n4 = ano4n['sexo_ignorado'].sum()

#Busca de todos os indivíduos mortos e nascidos de 2020
masculino_m5 = ano5m['sexo_masculino'].sum()
feminino_m5 = ano5m['sexo_feminino'].sum()
sexo_ignorado_m5 = ano5m['sexo_ignorado'].sum()
masculino_n5 = ano5n['sexo_masculino'].sum()
feminino_n5 = ano5n['sexo_feminino'].sum()
sexo_ignorado_n5 = ano5n['sexo_ignorado'].sum()


#Junção de todos os indivíduos mortos e nascidos de 2016
total_mortes1 = masculino_m1+feminino_m1+sexo_ignorado_m1
total_nasci1 = masculino_n1+feminino_n1+sexo_ignorado_n1

#Junção de todos os indivíduos mortos e nascidos de 2017
total_mortes2 = masculino_m2+feminino_m2+sexo_ignorado_m2
total_nasci2 = masculino_n2+feminino_n2+sexo_ignorado_n2

#Junção de todos os indivíduos mortos e nascidos de 2018
total_mortes3 = masculino_m3+feminino_m3+sexo_ignorado_m3
total_nasci3 = masculino_n3+feminino_n3+sexo_ignorado_n3

#Junção de todos os indivíduos mortos e nascidos de 2019
total_mortes4 = masculino_m4+feminino_m4+sexo_ignorado_m4
total_nasci4 = masculino_n4+feminino_n4+sexo_ignorado_n4

#Junção de todos os indivíduos mortos e nascidos de 2020
total_mortes5 = masculino_m5+feminino_m5+sexo_ignorado_m5
total_nasci5 = masculino_n5+feminino_n5+sexo_ignorado_n5

#União de todos os dados em arrays
Mortalidade = np.array([total_mortes1,total_mortes2,total_mortes3,total_mortes4,total_mortes5])
Natalidade = np.array([total_nasci1,total_nasci2,total_nasci3,total_nasci4,total_nasci5])
x = np.array([2016,2017,2018,2019,2020])

#Plot do gráfico
plt.title("Natalidade e Mortalidade ao Longo de 5 Anos", fontdict = font)
plt.xlabel("Anos", fontdict = font)
plt.ylabel("Dados em Milhões", fontdict = font)
plt.plot(x, Mortalidade, c = 'darkred', ls = '--', label = 'Mortalidade', linewidth = '4')
plt.plot(x, Natalidade, c = 'lightblue', label = 'Natalidade', linewidth = '4')
plt.grid()
plt.legend()
plt.show()
```
